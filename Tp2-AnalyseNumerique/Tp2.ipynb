### <font size="4"><font color="EA0A4E">**La méthode d'interpolation polynomiale, à quoi ça sert ?**

<font size="3">En analyse numérique, une fonction f **inconnue explicitement** est souvent connue seulement en certains points x0, x1, ..., xn 

Mais dans de nombreux cas, on a besoin d’effectuer des opérations (dérivation, intégration,
...) sur la fonction f.
    
On cherche donc à reconstruire f par une autre fonction fr **simple et facile à évaluer** à partir
des données discrètes de f.En espèrant que le modèle fr ne sera pas trop éloigné de la fonction
f aux autres points. 

### <font size="4"><font color="EA0A4E">**Explication de la méthode de lagrange**

<font size="3">Supposons qu'on à la courbe de f suivante:

<img src="image/courbef.PNG" />

##### <font color="03BF03"><font size="5">**Problème:** 

<font size="3">à partie de quelques points de f, on doit trouver une autre courbure qui passe par ces points à condition qu'elle se traduit par une expression d'un **polynôme**

##### <font color="03BF03"><font size="5">**La solution de Lagrange:**

<font size="3">Supposons qu'on à 3 points de coordonnées : **p1(x0,y0) p2(x1,y1) p2(x2,y2)**.
notre but c'est d'interpoler un polynome passant par p1,p2 et p3

<font size="4">**Démonstration:**

<font size="3">Dans cette démenstration, on va supposé que x0,x1 et x1 sont des **interrupteurs** et y0,y1 et y2 sont des **lampes** :

<img src="image/lagrange.PNG" />

<font size="3">Finalement, le polynome s'ecrit sous la forme : $P(x) = \dfrac{y0(x-x1)(x-x2)}{(x0-x1) + (x0-x2)} + \dfrac{y1(x-x0)(x-x2)}{(x1-x0) + (x1-x2)} + \dfrac{y2(x-x0)(x-x1)}{(x2-x0) + (x2-x1)}$

<font size="3">1)Que renvoie la fonction prog lorsqu'on lui déclare x=(x(1),x(2),...,x(n))x=(x(1),x(2),...,x(n)) et y=(y(1),y(2),...,y(n))y=(y(1),y(2),...,y(n)) deux vecteurs de même longueur n et c un nombre réel donné.

```python
import matplotlib . pyplot as plt
import numpy as np
import math
```

```python
from numpy import prod
def prog(x,y,c):
    S = [] 
    n = len(x)
    
    for j in range(n):
        L =[c-xx for xx in x]
        for k in range(n):
            if (k == j):                
                L[k] = 1
            else:
                L[k] = L[k]/(x[j] - x[k])   
        S.append(y[j]*prod(L))
        
    return  sum(S)
```

<font size="4"><font color="2233BF">La fonction **prog** ci-dessus permet de calculer le polynôme d'interpolation

```python
X=[1 ,2 ,3 ,4,-2];
Y=[-1, 0, 2 ,1,3];
for i in X:
    print(prog(X,Y,i))
sum(X)
```

<font size="4"><font color="2233BF">On remarque bien que lorsqu'on déclare les deux vecteurs **X=[1 ,2 ,3 ,4,-2] et Y=[-1, 0, 2 ,1,3 ]** , la fonction **prog** renvoie les images correspondantes de chaque point du vecteur.

<font size="4">1) Vérifier que sur $[-5,5]$ le polynôme d'interpolation de Lagrange de la fonction $f$ définie par $f(x)=\displaystyle{\frac{1}{1+x^2}}$, en $N$ points uniformément espacés ne convergent pas correctement aux voisinages de $-1$ et $1$, pour $N=7,8,9,10 ...$

```python
n=7
x=np.linspace(-5,5,n)
F=lambda x: 1/(1+x**2)
fig = plt.figure(1, figsize=(10, 8))
y=[F(xx) for xx in x]
p=np.polyfit(x,y,n-1)
t=np.linspace(-5,5,1000)
plt.subplot(221)
plt.plot(t,F(t),'r',x,y,'ob', t, np.polyval(p,t),'g')
plt.title('Interpolation Equidistante avec N=7')
plt.grid(True)
###############################################################################################################################
n=8
x=np.linspace(-5,5,n)
F=lambda x: 1/(1+x**2)
y=[F(xx) for xx in x]
p=np.polyfit(x,y,n-1)
t=np.linspace(-5,5,1000)
plt.subplot(222)
plt.plot(t,F(t),'r',x,y,'ob', t, np.polyval(p,t),'g')
plt.title('Interpolation Equidistante avec N=8')
plt.grid(True)
###############################################################################################################################
n=9
x=np.linspace(-5,5,n)
F=lambda x: 1/(1+x**2)
y=[F(xx) for xx in x]
p=np.polyfit(x,y,n-1)
t=np.linspace(-5,5,1000)
plt.subplot(223)
plt.plot(t,F(t),'r',x,y,'ob', t, np.polyval(p,t),'g')
plt.title('Interpolation Equidistante avec N=9')
plt.grid(True)
###############################################################################################################################
n=10
x=np.linspace(-5,5,n)
F=lambda x: 1/(1+x**2)
y=[F(xx) for xx in x]
p=np.polyfit(x,y,n-1)
t=np.linspace(-5,5,1000)
plt.subplot(224)
plt.plot(t,F(t),'r',x,y,'ob', t, np.polyval(p,t),'g')
plt.title('Interpolation Equidistante avec N=10')
plt.grid(True)
plt.show()
```

<font size="5"><font color="C4CC13">**Observation :**

<font size="4"><font color="2233BF">D'aprés ces courbes, on observe que le polynôme n'approche pas du tout f et si l'on avait plus de points (n augmente), toujours en les prenant de façon équirépartie (uniformément espacés), cela aurait été encore pire! On appelle ce phénomène de non-convergence des polynômes d'interpolation vers la fonction le **phénomène de Runge**.

<font size="4">2) Les points de Tchebycheff sur un intervalle $[a;b]$ sont définie par :$$x_k=\dfrac{a+b}{2}+\dfrac{a-b}{2}\cos\left(\frac{k-1}{n-1}\pi\right)\hspace{0.5cm} (k=1,..,n: x_1=a:x_n=b)$$

Vérifier que sur $[-5; 5]$ ; l'interpolation polynômiale de $f(x) = \dfrac{1}{1 + x^2}$ ; en $N$ points de Tchebychev est meilleure qu'une interpolation utilisant $N$ points uniforment espacés (on prendra $N=10, 11, 16, 17$).

```python
def tcheb(a,b,n):
    v=[]
    for k in range(n):
        v.append((a+b)/2+(a-b)/2*math.cos((k-1)/(n-1)*math.pi))
    return v
```

```python
tcheb(-5,5,11) #Test de la fonction tcheb
```

```python
n=10
F=lambda x:1/(1+x**2)
fig = plt.figure(1, figsize=(10, 8))
xt=tcheb(-7,7,n)
yt=[F(xx) for xx in xt]

pt=np.polyfit(xt,yt,n-1)
plt.subplot(221)
plt.plot(t,F(t),'r',xt,yt,'ob', t, np.polyval(pt,t),'g')
plt.title('Interpolation Tchebycheff avec N=10')
plt.grid(True)
###############################################################################################################################
n=11
F=lambda x:1/(1+x**2)
xt=tcheb(-7,7,n)
yt=[F(xx) for xx in xt]

pt=np.polyfit(xt,yt,n-1)
plt.subplot(222)
plt.plot(t,F(t),'r',xt,yt,'ob', t, np.polyval(pt,t),'g')
plt.title('Interpolation Tchebycheff avec N=11')
plt.grid(True)
###############################################################################################################################
n=16
F=lambda x:1/(1+x**2)
xt=tcheb(-7,7,n)
yt=[F(xx) for xx in xt]

pt=np.polyfit(xt,yt,n-1)
plt.subplot(223)
plt.plot(t,F(t),'r',xt,yt,'ob', t, np.polyval(pt,t),'g')
plt.title('Interpolation Tchebycheff avec N=16')
plt.grid(True)
###############################################################################################################################
n=17
F=lambda x:1/(1+x**2)
xt=tcheb(-7,7,n)
yt=[F(xx) for xx in xt]

pt=np.polyfit(xt,yt,n-1)
plt.subplot(224)
plt.plot(t,F(t),'r',xt,yt,'ob', t, np.polyval(pt,t),'g')
plt.title('Interpolation Tchebycheff avec N=17')
plt.grid(True)
plt.show()
```

<font size="4">1) Déterminer en utilisant les deux codes puis par un calcul à la main le polynôme de Newton pour les données suivantes :

```python
def intN(x,y):
    m=len(x)
    f=np.zeros(m)
    for k in range(1,m):
        f[k:m+1]=np.array(np.array(y[k:m+1])-np.array(y[k-1:m-1]))/np.array(np.array(x[k:m+1])-np.array(x[0:m-k]))
    return f
```

```python
x=[-1,0,2,5]
y=[6,1,3,66]
intN(x,y)
```

```python
def intNval(x,y,t):
    m=len(x)
    p=y[m-1]*np.ones(len(t))
    for k in range(m-2,-1,-1):
        p=p*np.array((np.array(t)-x[k]))+y[k]
    return p
```

```python
x=[-1,0,2,5]
y=[6,1,3,66]
t=[0,1,2,3]
intNval(x,y,t)
```
