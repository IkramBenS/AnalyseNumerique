```python
def horner(p, x):
     if len(p) == 1:
         return p[0]
     p[-2] += x * p[-1]
     return horner(p[:-1], x)
horner([1, -6, 15, -20, 15, -6, 1 ], 0.998)
```

<font size="5">**Question 2 :** 

Utiliser un programme dicho.m qui permettra de trouver les solutions de l'équation $f(x)=0$ avec $eps=0.001$

```python
def dicho(a,b,f,n):
    for i in range(n):
        c = (a+b)/2
        if f(a)*f(c) <= 0:
            b = c
        else:
            a = c
    return b
```

```python
x3=dicho(-2,-1.5, f, 10)
print(x3)
```

```python
x10=dicho(-2,-1.5,f,10)
x20=dicho(-2,-1.5,f,20)
print(x3,x10,x20)
```

<font size="5">**Question 3 :** 

Modifier le programme dicho.m pour qu'il donne le nombre d'itération nécessaire pour avoir la solution à $eps$ près.

```python
def dichow (a ,b, f , epsilon ):
    cp=0
    while (b-a) > epsilon :
        m = ( a + b )/2
        if f(m) == 0 :
            return m
        if f(a)*f(m) <0 :
            b=m
        else :
            a = m
        cp+=1
    print (" nombre d'iterations necessaires :", cp )
    return m, cp
```

```python
dichow(-2,-1.5,f,0.0001)
```

```python
def newton(f,fprime,a,eps):
    x1=a
    x2=x1-(f(x1)/fprime(x1))
    while abs(x1-x2)>eps :
        x1=x2
        x2=x1-(f(x1)/fprime(x1))
        m=x2
    return m
```

On va applique l'algorithme Newton à f(x) avec : **𝑓(𝑥)=𝑒−𝑥−𝑥** , en prenant  𝑥0=0 ,  𝑒𝑝𝑠=0.001 

```python
import math
from sympy import *
from sympy.abc import x

f=lambda x: math.exp(-x)-x
fprime=lambda x: -math.exp(-x)-1
newton(f,fprime,0,0.001)
```

```python
newton(f,fprime,1,0.001)
```

Donc l'équation  𝑓(𝑥)=0  admet une solution unique que l'on peut obtenir en utilisant la méthode de Newton à partir >de  𝑥0=1 .
